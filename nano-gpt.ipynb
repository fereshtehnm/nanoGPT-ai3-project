```python
# read & inspect the dataset
with open('input.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
print(f"length of dataset in characters: {len(text):,}")
print(text[:1000])
```

```python
characters = sorted(list(set(text)))
vocab_size = len(characters)
print("all the unique characters:", ''.join(characters))
print(f"vocab size: {vocab_size:,}")
```

```python
stoi = { ch:i for i,ch in enumerate(characters) }
itos = { i:ch for i,ch in enumerate(characters) }
encoding = lambda s: [stoi[c] for c in s] # encoder: take a string, output a list of integers
decoding = lambda l: ''.join([itos[i] for i in l]) # decoder
print(encoding("hii there"))
print(decoding(encoding("hii there")))
```
